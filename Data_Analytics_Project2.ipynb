```python
import matplotlib.pyplot as plt

df.hist(figsize=(12,8))
plt.tight_layout()
plt.show()
```

```python
numeric_cols = ['Quantity','UnitPrice','ItemsInCart','TotalPrice']

for col in numeric_cols:
    plt.figure(figsize=(6,4))
    plt.boxplot(df[col])
    plt.title(col)
    plt.show()
```

```python
df['Product'].value_counts().plot(kind='bar', figsize=(8,5))

plt.title("Orders by Product")

plt.xlabel("Product")

plt.ylabel("Count")

plt.show()
```

```python
df['PaymentMethod'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(6,6))

plt.ylabel("")

plt.title("Payment Method Distribution")

plt.show()
```

```python
top_products = df.groupby('Product')['TotalPrice'].sum().sort_values(ascending=False)

top_products.head(10).plot(kind='bar', figsize=(10,5))

plt.title("Top 10 Products by Revenue")
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.xticks(rotation=45)

plt.show()
```

```python
customer_spending = df.groupby('CustomerID')['TotalPrice'].sum()

segments = pd.qcut(
    customer_spending,
    q=4,
    labels=['Low', 'Medium', 'High', 'Premium']
)

segment_df = pd.DataFrame({
    'TotalSpending': customer_spending,
    'Segment': segments
})

segment_df.head()
```

```python
segment_df['Segment'].value_counts().plot(
    kind='bar',
    figsize=(8,5)
)

plt.title("Customer Segments")
plt.ylabel("Number of Customers")
plt.show()
```

```python
coupon_analysis = df.groupby('CouponCode')['TotalPrice'].agg(['count', 'mean'])

coupon_analysis
```

```python
basket = df.groupby('Product')['ItemsInCart'].mean()

basket.sort_values(ascending=False)
```

```python
product_sales = df.groupby('Product')['TotalPrice'].sum()

revenue_percent = (product_sales / product_sales.sum()) * 100

revenue_percent.sort_values(ascending=False)
```

```python
clv = df.groupby('CustomerID').agg({
    'TotalPrice':'sum',
    'OrderID':'count'
})

clv.columns = ['LifetimeValue','Orders']

clv['AverageOrderValue'] = clv['LifetimeValue'] / clv['Orders']

clv.head()
```

```python
shipping = df.groupby('OrderStatus').agg({
    'TotalPrice':['count','sum']
})

shipping
```

```python
dashboard = df.groupby('Product').agg({
    'Quantity':'sum',
    'TotalPrice':'sum',
    'OrderID':'count'
})

dashboard.columns = ['Units Sold','Revenue','Orders']

dashboard.sort_values('Revenue', ascending=False)
```

```python
dashboard.to_csv("Product_Performance_Report.csv", index=True)

print("Report saved successfully!")
```

```python
top_customers = df.groupby('CustomerID')['TotalPrice'].sum().nlargest(5)

print(top_customers)
```

```python
highest_product = df.groupby('Product')['TotalPrice'].sum().idxmax()

print("Highest Revenue Product:", highest_product)
```

```python
payment = df['PaymentMethod'].mode()[0]

print("Most Preferred Payment Method:", payment)
```
